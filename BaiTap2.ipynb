```python
import pandas as pd
import geopandas
```

```python
huyen = geopandas.read_file("data/Bai2_Chuyen doi he TD/HUYEN_region.shp")
huyen.plot()
```

```python
huyen.crs
```

```python
huyen_mercator = huyen.to_crs(epsg='4326')
huyen_mercator.plot()
```

```python
import folium
```

```python
map = folium.Map()
```

```python
folium.GeoJson(huyen_mercator.to_json()).add_to(map)
folium.LayerControl().add_to(map)
map
```
